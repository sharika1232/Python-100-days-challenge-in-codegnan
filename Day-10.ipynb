```python
# keyword arguments
# multiplication of two numbers 
a = int(input("Enter a number 1:"))
b = int(input("Enter a number 2:"))
def multiplication(n1,n2):
    res = n1 * n2
    return res
print(multiplication(n1=a,n2=b))
print(multiplication(n2=b,n1=a))
```

```python
# practice
# 1. Sum of two numbers using functions
def sum_of_two_numbers(n1,n2):
    res = n1 + n2
    return res
res = sum_of_two_numbers(20,30)
print(res)
```

```python
#2. sum of 4 numbers usig functions
def Adding_four_numbers(z1,z2,z3,z4):
    res = z1 + z2 + z3 + z4
    return res
a = int(input("Enter a number 1:"))
b = int(input("Enter a number 2:"))
c = int(input("Enter a number 3:"))
d = int(input("Enter a number 4:"))
res = Adding_four_numbers(a,b,c,d)
print(res)
```
