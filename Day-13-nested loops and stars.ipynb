```python
for i in range(1,4):
    for j in range(1,11):
        print(f"{i} * {j} = {i*j}")
    print("________________")
```

```python
def tables(n,o):
    for i in range(1,11):
        print(f"{n} * {i} = {n*i}")
    for j in range(1,11):
        print(f"{o} * {j} = {o*j} ")
tables(5,4)
    
```

```python
# print n starts in horizontally
n = int(input())
for i in range(1, n+1):
    print("*", end = ' ')
```
