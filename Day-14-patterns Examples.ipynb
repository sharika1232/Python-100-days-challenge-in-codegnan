```python
n = 4
k = 1
for i in range(1, n+1):
    for j in range(1, n+1):
        if i >= j:
            print(k, end = " ")
            k+=1 
        else:
            print(" ", end= " ")
    print()
```

**Hallow  square** 

```python
n = 6
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end =" ")
        else:
            print(" ", end = " ")
    print()
```
